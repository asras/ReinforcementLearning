```python
env = BlackjackEnv()
```

```python
def make_epsilon_greedy_policy(Q, epsilon, nA):
    """
    Creates an epsilon-greedy policy based on a given Q-function and epsilon.
    
    Args:
        Q: A dictionary that maps from state -> action-values.
            Each value is a numpy array of length nA (see below)
        epsilon: The probability to select a random action . float between 0 and 1.
        nA: Number of actions in the environment.
    
    Returns:
        A function that takes the observation as an argument and returns
        the probabilities for each action in the form of a numpy array of length nA.
    
    """
    def policy_fn(observation):
        randomNum = np.random.random()
        if (randomNum >= epsilon):
            return np.argmax(Q[observation])
        else:
            return np.random.randint(nA)
    return policy_fn
```

```python
def mc_control_epsilon_greedy(env, num_episodes, discount_factor=1.0, epsilon=0.1):
    """
    Monte Carlo Control using Epsilon-Greedy policies.
    Finds an optimal epsilon-greedy policy.
    
    Args:
        env: OpenAI gym environment.
        num_episodes: Nubmer of episodes to sample.
        discount_factor: Lambda discount factor.
        epsilon: Chance the sample a random action. Float betwen 0 and 1.
    
    Returns:
        A tuple (Q, policy).
        Q is a dictionary mapping state -> action values.
        policy is a function taht takes an observation as an argument and returns
        action probabilities
    """
    
    # Keeps track of sum and count of returns for each state
    # to calculate an average. We could use an array to save all
    # returns (like in the book) but that's memory inefficient.
    returns_sum = defaultdict(float)
    returns_count = defaultdict(float)
    
    # The final action-value function.
    # A nested dictionary that maps state -> (action -> action-value).
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    
    # The policy we're following
    policy = make_epsilon_greedy_policy(Q, epsilon, env.action_space.n)
    
    # Implement this!
    
    for _ in range(num_episodes):
        ## run an episode using policy
        episode = []
        observation = env.reset()
        while (True):            
            action = policy(observation)
            previousObservation = observation
            observation, reward, done, _ = env.step(action)
            episode.append((previousObservation, action, reward))
            if done:
                break
                
        ##Use observations to update policy
        stateActions_in_episode = set(tuple((x[0], x[1])) for x in episode)
        for stateAction in stateActions_in_episode:
            first_occurrence = next(i for i, x in enumerate(episode) if tuple((x[0],x[1])) == stateAction)
            G = sum(x[2]*(discount_factor**i) for i,x in enumerate(episode[first_occurrence:]))
            returns_sum[stateAction] += G
            returns_count[stateAction] += 1
            Q[stateAction[0]][stateAction[1]] = returns_sum[stateAction]/returns_count[stateAction]
        
        
        ##Shouldnt need this, haven't tested
        policy = make_epsilon_greedy_policy(Q, epsilon, env.action_space.n)
    
    ## repeat
    
    return Q, policy
```

```python
Q, policy = mc_control_epsilon_greedy(env, num_episodes=500000, epsilon=0.1)
```

```python
# For plotting: Create value function from action-value function
# by picking the best action at each state
V = defaultdict(float)
for state, actions in Q.items():
    action_value = np.max(actions)
    V[state] = action_value
plotting.plot_value_function(V, title="Optimal Value Function")
```
